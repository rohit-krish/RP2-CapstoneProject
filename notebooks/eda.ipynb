```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

plt.style.use(plt.style.available[11])
```

```python
df = pd.read_csv('./data/preprocessed_data.csv')
df.head()
```

## 1. which is the common genre?


```python
df['genre'].value_counts().plot(kind='barh')
plt.show()
```

## 2. which genre has the most score?


```python
df.groupby('genre')['score'].mean().sort_values(ascending=False).plot(kind='barh', logx=True)
plt.show()
```

## 3. which director, writer, star has the most score?


```python
df.groupby('director')['score'].max().sort_values(ascending=False)[:10].plot(kind='barh', logx=True)
plt.show()

df.groupby('writer')['score'].max().sort_values(ascending=False)[:10].plot(kind='barh', logx=True)
plt.show()

df.groupby('star')['score'].max().sort_values(ascending=False)[:10].plot(kind='barh', logx=True)
plt.show()
```

## 4. which is the most common rating?


```python
df['rating'].value_counts().plot(kind='pie')
plt.show()
```

## 5. which rating has the most score?


```python
df.groupby("rating")["score"].max().plot(kind='bar',logy=True)
plt.show()
```

## 6. which is the most common genre in each year?

```python
df.groupby('year')['genre'].value_counts().groupby('year').idxmax().apply(lambda x: x[1])
```
