```python
def get_year(string):
    pattern = r'\b\d{4}\b'
    match = re.search(pattern, string)

    if match:
        return match.group()
    else:
        return np.NaN
```

```python
def get_country(string):
    pattern = r'\(([\w ]+)\)'
    res = re.findall(pattern, string)

    if len(res) == 0:
        return np.NaN
    else:
        return res[0]
```

```python
print(get_year('October 20, 2000 (United States)'))
print(get_country('October 20, 2000 (United States)'))
```

```python
df['year'] = df['released'].apply(get_year)
df['country'] = df['released'].apply(get_country)

df = df.drop('released', axis=1)

df.head()
```
